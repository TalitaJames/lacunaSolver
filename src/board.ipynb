```python
# https://stackoverflow.com/questions/52447723/networkx-creating-a-complete-graph-for-a-given-set-of-nodes
import itertools
nodeIDs = list(range(len(tokens))) # get numbers 0 - size of tokens
edges = itertools.combinations(nodeIDs, 2)
# print(*edges)
T.add_edges_from(edges) 
```

```python
# Generate list of edges only in same colour:
for i in range(7):
    nodeIDs = [ n for n,atribute in T.nodes().items() if atribute['type'] == i] 
    edges = itertools.combinations(nodeIDs, 2)
    T.add_edges_from(edges) 
```
